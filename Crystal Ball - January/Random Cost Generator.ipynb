```python
import pandas as pd
import numpy as np
df = pd.read_csv('JanWeeklyDemand.csv')
```

```python
items = df['ItemCode'].str.lstrip('0')
items = items.unique()
partlist=[]
for p in items:
    if p[0:2] == 'SR':
        p2 = p[2:]
    else:
        p2 = p
    partlist.append(p2)
items = pd.Series(partlist).unique()
```

```python
branches = pd.read_csv('JanBranches.csv')
```

```python
branches = branches['Branch'].values.tolist()
```

```python
tuplist = []
for j in branches:
    for k in branches:
        for i in items: 
            tuplist.append((j,k,i))
    
```

```python
data = pd.DataFrame(tuplist,columns = ['From','To','Item'])
```

```python
data['Cost'] = np.random.randint(1,101,size = len(data))
data = data.groupby(by = ['From','To','Item']).mean()
```
